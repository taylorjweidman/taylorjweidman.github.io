# Normal Distribution

## Find the value at a percentile

```python
from scipy.stats import norm

# Normal distribution with mean 10 and standard deviation 2
mean = 10
std_dev = 2
custom_normal = norm(loc=mean, scale=std_dev)

# 16th percentile of a normal distribution with mean 10 and standard deviation 2
value_16th_custom = custom_normal.ppf(0.16)
print("16th percentile for N(10, 2):", value_16th_custom)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

percentile_example = 0.16

# Define parameters for the custom normal distribution
mean = 10
std_dev = 2
custom_normal = norm(loc=mean, scale=std_dev)

# Calculate the 16th percentile
value_16th_custom = custom_normal.ppf(percentile_example)
print("16th percentile for N(10, 2):", value_16th_custom)

# Generate x values for plotting
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = custom_normal.pdf(x)

# Plot the normal distribution curve
plt.plot(x, y, label='N(10, 2) Distribution', linewidth=2)

# Highlight the area up to the 16th percentile
x_fill = np.linspace(mean - 4*std_dev, value_16th_custom, 1000)
y_fill = custom_normal.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='lightcoral', alpha=0.4, label='16th Percentile Area')

# Mark the 16th percentile
plt.axvline(value_16th_custom, color='red', linestyle='--', label=f'16th Percentile ({value_16th_custom:.2f})')

# Labeling
plt.title("Normal Distribution N(10, 2) with 16th Percentile Highlighted")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()

# Display the plot
plt.show()
```

## Find the area between two values

```python
from scipy.stats import norm

# Define the mean and standard deviation of the distribution
mu = 0     # Mean of the distribution
sigma = 1  # Standard deviation of the distribution

# Define the values a and b
a = -1
b = 1

# Calculate the cumulative distribution function (CDF) at a and b
P_a = norm.cdf(a, mu, sigma)  # Probability of X <= a
P_b = norm.cdf(b, mu, sigma)  # Probability of X <= b

# The probability of X being between a and b
probability = P_b - P_a

print(f"The probability that X is between {a} and {b} is: {probability}")
```

## Plot the area between two values

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0     # Mean
sigma = 1  # Standard deviation
a = -1     # Lower bound
b = 2      # Upper bound

# Generate x values for the normal distribution curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution', linewidth=2)

# Fill the area between a and b
x_fill = np.linspace(a, b, 1000)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.4, label=f'P({a} ≤ X ≤ {b})')

# Labeling
plt.title("Normal Distribution with Highlighted Area between a and b")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()

# Display the plot
plt.show()
```

## Example 1 | Q1

```python
from scipy.stats import norm

# Define the parameters of the normal distribution
mean = 18    # Mean waiting time in minutes
std_dev = 4  # Standard deviation of waiting time in minutes

# Define the time we're interested in (less than 6 minutes)
time = 6

# Calculate the probability using the CDF
probability = norm.cdf(time, mean, std_dev)

print(f"The probability that a customer waits less than {time} minutes is: {probability:.6f}")
```

## Example 2 | Q2

```python
from scipy.stats import norm

# Define the z-value
z_value = -2

# Calculate the probability of z > -2
# This is equivalent to 1 - P(Z <= -2)
probability = 1 - norm.cdf(z_value)

print(f"The probability of a z-value being greater than {z_value} is: {probability:.4f}")
```

## t-Distribution

```python
import numpy as np
from scipy import stats

# Sample data
data = [5.3, 5.7, 5.8, 6.1, 5.4, 5.5, 5.6, 5.9, 6.0]
sample_size = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

# Confidence level
confidence = 0.95
alpha = 1 - confidence
df = sample_size - 1  # degrees of freedom

# Calculate the t critical value for the given confidence level and df
t_critical = stats.t.ppf(1 - alpha/2, df)

# Calculate the margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Calculate the confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean}")
print(f"{confidence*100}% Confidence Interval: ({ci_lower}, {ci_upper})")
```
