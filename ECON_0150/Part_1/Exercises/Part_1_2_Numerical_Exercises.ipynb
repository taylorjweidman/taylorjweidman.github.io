## Discrete Data 

### Summary and Visualization

This notebook summarizes and visualizes the discrete numerical dataset.

### Load the discrete dataset

```python
discrete_data_path = "discrete_numerical_dataset.csv"  # Update this path if necessary
discrete = pd.read_csv(discrete_data_path)
```

### Summarize the dataset

```python
discrete_summary = discrete['Number of Children'].value_counts().sort_index()
print(discrete_summary)
```

```python
discrete.describe()
```

### Visualize the dataset

```python
plt.figure(figsize=(8, 6))

bins = discrete_summary.index
plt.hist(discrete['Number of Children'], bins=bins, edgecolor='white')

plt.title("Distribution of Number of Children")
plt.xlabel("Number of Children")
plt.ylabel("")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

## Continuous Data 

### Summary and Visualization

This notebook summarizes and visualizes the continuous numerical dataset.

### Load the continuous dataset

```python
continuous_data_path = "continuous_numerical_dataset.csv"  # Update this path if necessary
continuous = pd.read_csv(continuous_data_path)
```

### Summarize the dataset

```python
continuous_summary = continuous['Household Income (USD)'].describe()
print(continuous_summary)
```

### Visualize the dataset - Histogram

```python
plt.figure(figsize=(8, 6))
continuous['Household Income (USD)'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Household Income")
plt.xlabel("Household Income (USD)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Visualize the dataset - Boxplot

```python
plt.figure(figsize=(8, 6))
continuous['Household Income (USD)'].plot(kind='box', vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Household Income")
plt.xlabel("Household Income (USD)")
plt.show()
```

### Visualize the dataset - Boxplot with horizontal scatter


```python
plt.figure(figsize=(8, 6))
plt.boxplot(continuous['Household Income (USD)'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.scatter(continuous['Household Income (USD)'], [1] * len(continuous), alpha=0.5, color='darkblue', label='Data Points')
plt.title("Boxplot of Household Income with Data Points")
plt.xlabel("Household Income (USD)")
plt.legend()
plt.show()
```

### Visualize the dataset - Boxplot with jittered horizontal scatter 

```python
plt.figure(figsize=(8, 3))
plt.boxplot(continuous['Household Income (USD)'], vert=False, patch_artist=True, boxprops=dict(facecolor='royalblue', alpha=0.5), zorder=2)

import numpy as np
jitter = np.random.uniform(0.9,1.1, len(continuous))

plt.scatter(continuous['Household Income (USD)'], jitter, alpha=0.1, color='red', label='Data Points', zorder=3)
plt.title("Boxplot of Household Income with Data Points")
plt.xlabel("Household Income (USD)")
plt.ylabel('')
plt.yticks([])
plt.legend()
plt.show()
```

### Visualize the dataset - Boxplot with jittered horizontal scatter and hist

```python
import matplotlib.gridspec as gridspec

# Visualize the dataset - Combined Boxplot and Histogram
fig = plt.figure(figsize=(8, 4))
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1])

# Boxplot
ax0 = plt.subplot(gs[0])
ax0.boxplot(continuous['Household Income (USD)'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

jitter = np.random.uniform(0.9,1.1, len(continuous))

ax0.scatter(continuous['Household Income (USD)'], jitter, alpha=0.5, color='darkblue', label='Data Points', zorder=3)
ax0.set_title("Boxplot of Household Income with Data Points")
ax0.set_xlabel("")
ax0.legend()
ax0.grid(axis='x', linestyle='--', alpha=0.7)

# Histogram
ax1 = plt.subplot(gs[1])
continuous['Household Income (USD)'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black', ax=ax1)
ax1.set_title("Distribution of Household Income")
ax1.set_xlabel("Household Income (USD)")
ax1.set_ylabel("Frequency")
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
