## Ordinal Data Summary and Visualization

This notebook summarizes and visualizes the ordinal categorical dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the ordinal dataset

```python
data_path = "ordinal_categorical_dataset.csv"  # Update this path if necessary
df = pd.read_csv(data_path)
```

### Summarize the dataset

```python
summary = df['Economic Optimism'].value_counts().sort_index()
print(summary)
```

### Visualize the dataset

```python
plt.figure(figsize=(8, 6))
summary.plot(kind='barh', color=['purple', 'blue', 'green', 'orange', 'red'])
plt.title("Economic Optimism Distribution")
plt.xlabel("Count")
plt.ylabel('')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
