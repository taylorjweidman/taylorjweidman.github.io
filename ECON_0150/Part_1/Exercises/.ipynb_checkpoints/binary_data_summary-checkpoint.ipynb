## Binary Data Summary and Visualization

This notebook summarizes and visualizes the binary categorical dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the binary dataset

```python
data_path = "binary_categorical_dataset.csv"  # Update this path if necessary
df = pd.read_csv(data_path)
```

### Summarize the dataset

```python
summary = df['Employment Status'].value_counts()
print(summary)
```

### Visualize the dataset

```python
plt.figure(figsize=(8, 6))
summary.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Employment Status Distribution")
plt.xlabel("Employment Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
